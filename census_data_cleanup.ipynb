```
import pandas as pd
import matplotlib.pyplot as plt
import folium
```

```
df_pop1 = pd.read_csv('/content/census_neighborhood_demographics_2010.csv')
df_pop1.head()
```

```
df_pop1_area = df_pop1.loc[:, ('NBRHD_NAME', 'POPULATION_2010', 'HOUSINGUNITS_2010', 'MALE', 'FEMALE')]
df_pop1_area.head()
```

```
df_pop1_area.rename({'NBRHD_NAME':'neighborhood', 'POPULATION_2010':'population', 'HOUSINGUNITS_2010':'num_houses'}, axis=1, inplace=True)
df_pop1_area.head()
```

```
df_pop1_area.sort_values(by=['population'], axis=0, ascending = False, inplace=True)
df_pop1_area.set_index('neighborhood', drop=True, inplace=True)
df_pop1_area.head()
```

```
df_pop_age1 = df_pop1.loc[:, ('NBRHD_NAME', 'AGE_0_TO_9', 'AGE_10_TO_19', 'AGE_20_TO_29')]
df_pop_age2 = df_pop1.loc[:, ('NBRHD_NAME', 'AGE_30_TO_39', 'AGE_40_TO_49', 'AGE_50_TO_59')]
df_pop_age3 = df_pop1.loc[:, ('NBRHD_NAME', 'AGE_60_TO_69', 'AGE_70_TO_79', 'AGE_80_PLUS' )]
```

```
df_pop_age1.set_index('NBRHD_NAME', drop=True, inplace=True)
df_pop_age2.set_index('NBRHD_NAME', drop=True, inplace=True)
df_pop_age3.set_index('NBRHD_NAME', drop=True, inplace=True)
```

```
df_pop_age2
df_pop_age3
df_pop_age1.head()
```

```
df_age1 = df_pop_age1.sum(axis=1)
df_age2 = df_pop_age2.sum(axis=1)
df_age3 = df_pop_age3.sum(axis=1)
```

```
df_pop1_area['under30'] = df_age1
df_pop1_area['30-60'] = df_age2
df_pop1_area['over60'] = df_age3
```

```
df_pop1_area.head()
```

```
df_pop1_area.to_csv('/content/census_cleaned.csv')
```
